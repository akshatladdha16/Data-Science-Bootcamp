# Asking and Answering Questions

```python
# Q1 from which locality max hotels are listed
delhi=df2[(df2.City=="New Delhi")]
plt.figure(figsize=(12,6))
sns.barplot(x=delhi.Locality.value_counts().head(10),y=delhi.Locality.value_counts().head(10).index)

plt.ylabel(None)
plt.xlabel("Number of restaurants")
plt.title("Restaurants Listing on Zomato in New Delhi")
```

```python
delhi
```

Connaught place having maximum listed restaurants and then Rajouri Garden

```python
# Q2 fetching restaurants with having "Excellent" and "Very good" rating 
cp= delhi[(delhi.Locality.isin(['Connaught Place'])) & (delhi['Rating text'].isin(['Excellent','Very Good']))]
cp
```

```python
# Q3 what Cuisines do these top rated restaurants offer
cuisines_cp=cp.Cuisines.value_counts().reset_index()
# now extracting the cuisines in sigle list
cuisien=[]
for x in cuisines_cp['index']:
    cuisien.append(x)
cuisien
```

```python
# now to visualise the info about what is most frequent cuisien is we will be using WordCloud
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
  
comment_words = ''
stopwords = set(STOPWORDS)
  
# iterate through the csv file
for val in cuisien:
      
    # typecaste each val to string
    val = str(val)
  
    # split the value
    tokens = val.split()
      
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 1500, height = 1500,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
  
# plot the WordCloud image                       
plt.figure(figsize = (8, 8), facecolor = 'orange')
plt.title('Resturants cuisien -  Top Resturants')
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
  
plt.show()
```
